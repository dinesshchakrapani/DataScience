<p><b>3.</b> What errors are defined in the ArithmeticError class? Explain any two with an example.<br>

<b>Ans:</b> Errors occur during arithmetic operations is defined inside this ArithmeticError Class.<br><br>
<b>ZeroDivisionError:</b> This is an exception that is raised when an attempt is made to divide a number by zero. <br>
<b>Code:</b>

```python
x = 10
y = 0

try:
    z = x / y
except ZeroDivisionError:
    print("Error: division by zero")
```

<b>OverflowError:</b> This is an exception that is raised when the result of an arithmetic operation is too large to be represented by the system's integer type.<br>
<b>Code:</b>

```python
try:
    for i in range(1, 1000):
        j = j**i
except OverflowError:
    print("Error: integer overflow")
```

<p><b>4.</b> Why LookupError class is used? Explain with an example KeyError and IndexError.<br>

<b>Ans:</b> Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists. The two types of exceptions raised are IndexError, KeyError.<br><br>
    <b>Key Error:</b> This occurs when we try to access the key which is not present in the dictonary<br>
    <b>Code:</b>
<p>


```python
try:
    a={1:"din",2:"lin"}
    a[3]
except KeyError:
    print("Error: Key not found in the dictonary")
```

<b>Index Error:</b> This occurs when we try to access the index which is not present in the list<br>
    <b>Code:</b>

```python
try:
    a=[1,2,3]
    a[3]
except IndexError:
    print("Error: Index not found in the list")
```

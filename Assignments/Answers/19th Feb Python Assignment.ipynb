```python
'''
Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

Code:
'''
from flask import Flask

app = Flask(__name__)
@app.route("/")
def hello_world():
    return "Hello World!!"
if __name__=="__main__":
    app.run(host="0.0.0.0")

```

```python
'''
Q3. What is App routing in Flask? Why do we use app routes?

Ans:App routing refers to mapping URLs (Uniform Resource Locators) to view functions. In other words, when a user makes a 
request to a particular URL, Flask needs to know which function to call in response to that request. This is done using app 
routing.App routing in Flask is defined using the @app.route decorator. This decorator is used to bind a URL to a view function.

We use app routes in Flask to create a structured URL scheme for our web application. This makes it easy for users to navigate 
the application and for search engines to index the content. In addition, app routes help to keep our code organized and 
modular, making it easier to maintain and update our application over time.
'''
```

```python
'''
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

Code:
'''
from flask import Flask
app = Flask(__name__)
dic={"Company_Name": "ABC Corporation",
"Location": "India",
"Contact_Detail": "999-999-9999"}
@app.route("/")
def hello_world():
    a=""
    for i in dic:
        a+=i+": "+dic[i]+"<br>"
    return a
@app.route("/welcome")
def welcome():
    return f"Welcome to {dic['Company_Name']}"
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
'''
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans: The url_for() function is used for URL building. It generates a URL for the specified endpoint with the given arguments.

Code:
'''
from flask import Flask,url_for

app = Flask(__name__)
@app.route('/')
def index():
    url = url_for('hello')
    return f'The URL for the hello() function is: {url}'

@app.route('/hello_world')
def hello():
    return 'Hello, World!'
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
